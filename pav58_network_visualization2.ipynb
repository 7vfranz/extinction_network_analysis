```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import os
os.chdir('/Users/franzv/Desktop/dataSci/ds_projects/Pav58_Network')
os.getcwd()
```

```python
brain_node_labels = ['IL','PL','mOFC',
                    'lOFC','vOFC','NAcC',
                    'mNAcSh','lNAcSh','aPVT',
                    'mPVT','pPVT','BLA']

node_colors=['#31810D','#3A765E','#FC66FE',
            '#992B54','#8037A2', '#FC763B',
            '#FB400C','#FC6665','#0091ea',
            '#3469A8','#7696FB','#949494']

p_recall_nodesize=[3900,4400,3900,3400,3400,4900,6400,4400,3900,4400,4400,2400]
up_recall_nodesize=[3400,4400,3900,3400,3400,4900,5900,2400,2400,2400,3400,2400]
p_ext_nodesize=[3900,3400,3900,2400,2400,3900,4900,3900,3400,3400,3400,2400]
up_ext_nodesize=[3900,3400,3900,2400,2400,3900,4900,4400,3400,2400,3400,2400]
hc_nodesize=[2400,2400,2400,2400,2400,2400,2400,2400,2400,2400,2400,2400]

hc_nodesize_small = [x//5 for x in hc_nodesize]
p_recall_nodesize_small = [x//5 for x in p_recall_nodesize]
p_ext_nodesize_small = [x//5 for x in p_ext_nodesize]
up_recall_nodesize_small = [x//5 for x in up_recall_nodesize]
up_ext_nodesize_small = [x//5 for x in up_ext_nodesize]

w1 = 0.25
w2 = 1.25
w3 = 3.25
```

```python
#P Recall Graph

plt.figure(1, figsize=(7,7), dpi=300)

G1 = nx.Graph()
G1.add_nodes_from(brain_node_labels)

G1.add_weighted_edges_from([('IL','PL',w2),('IL','mOFC',w2),
                            ('IL','lOFC',w2),('IL','vOFC',w3),('IL','pPVT',w1)])

G1.add_weighted_edges_from([('PL','lOFC',w2),('PL','vOFC',w1)])

G1.add_weighted_edges_from([('mOFC','vOFC',w2),('mOFC','aPVT',w1),
                            ('mOFC','mPVT',w1),('mOFC','pPVT',w1)])

G1.add_weighted_edges_from([('lOFC','vOFC',w2)])

G1.add_weighted_edges_from([('vOFC','pPVT',w1)])

G1.add_weighted_edges_from([('NAcC','mNAcSh',w2)])

G1.add_weighted_edges_from([('mNAcSh','lNAcSh',w3),('mNAcSh','pPVT',w1)])

G1.add_weighted_edges_from([('lNAcSh','pPVT',w2)])

G1.add_weighted_edges_from([('aPVT','mPVT',w3)])

G1.add_weighted_edges_from([('mPVT','pPVT',w1)])

G1.add_weighted_edges_from([('pPVT','BLA',w1)])

weights=nx.get_edge_attributes(G1,'weight').values()

pos = nx.circular_layout(G1, scale=0.8)
nx.draw(G1, pos=pos, with_labels = False,
        node_color='gray',
       node_size=p_recall_nodesize_small,
       width=list(weights))


pos_nodes = nx.circular_layout(G1, scale=1.4)
nx.draw_networkx_labels(G1, pos_nodes, font_size=28)

plt.margins(x=0.61,y=0.61)
plt.show()
```

```python
#P Ext 

plt.figure(2, figsize=(7,7), dpi=300)

G2 = nx.Graph()
G2.add_nodes_from(brain_node_labels)

G2.add_weighted_edges_from([('IL','PL',w3),('IL','mOFC',w3),
                                ('IL','vOFC',w1),('IL','NAcC',w1),
                                ('IL','lNAcSh',w1),('IL','aPVT',w1),
                                ('IL','mPVT',w1),('IL','pPVT',w2),
                                ('IL','BLA',w1)])

G2.add_weighted_edges_from([('PL','mOFC',w3),('PL','vOFC',w1),
                                ('PL','NAcC',w2),('PL','lNAcSh',w2),
                                ('PL','aPVT',w2),('PL','mPVT',w1),
                                ('PL','pPVT',w2),('PL','BLA',w1)])

G2.add_weighted_edges_from([('mOFC','vOFC',w1),('mOFC','lNAcSh',w1),('mOFC','pPVT',w1)])

G2.add_weighted_edges_from([('lOFC','vOFC',w3),('lOFC','NAcC',w2),
                                ('lOFC','mNAcSh',w2),('lOFC','lNAcSh',w3),
                                ('lOFC','aPVT',w1),('lOFC','mPVT',w1),('lOFC','BLA',w1)])

G2.add_weighted_edges_from([('vOFC','NAcC',w2),('vOFC','mNAcSh',w1),
                                ('vOFC','lNAcSh',w3),('vOFC','aPVT',w1),('vOFC','BLA',w1)])

G2.add_weighted_edges_from([('NAcC','mNAcSh',w3),('NAcC','lNAcSh',w3),
                                ('NAcC','aPVT',w3),('NAcC','mPVT',w3),
                                ('NAcC','pPVT',w2),('NAcC','BLA',w2)])

G2.add_weighted_edges_from([('mNAcSh','lNAcSh',w3),('mNAcSh','aPVT',w3),
                                ('mNAcSh','mPVT',w2),('mNAcSh','pPVT',w1),('mNAcSh','BLA',w1)])

G2.add_weighted_edges_from([('lNAcSh','aPVT',w2),('lNAcSh','mPVT',w2),
                           ('lNAcSh','pPVT',w2),('lNAcSh','BLA',w1)])

G2.add_weighted_edges_from([('aPVT','mPVT',w3),('aPVT','pPVT',w3)])

G2.add_weighted_edges_from([('mPVT','pPVT',w3)])

weights=nx.get_edge_attributes(G2,'weight').values()

pos = nx.circular_layout(G2, scale=0.8)
nx.draw(G2, pos=pos, with_labels = False,
        node_color='gray',
       node_size=p_ext_nodesize_small,
       width=list(weights))

pos_nodes = nx.circular_layout(G2, scale=1.4)
nx.draw_networkx_labels(G2, pos_nodes, font_size=28)

plt.margins(x=0.61,y=0.61)
# plt.savefig('paired_ext3.png', bbox_inches='tight')
plt.show()

```
